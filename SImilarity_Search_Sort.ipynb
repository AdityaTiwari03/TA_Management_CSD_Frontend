```python
import pandas as pd
df = pd.read_csv('/Book1.csv')
df["combined"] =  df.apply(lambda row: ' '.join(row.values.astype(str)), axis=1)
```

```python
from openai import OpenAI
client = OpenAI(api_key="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")

def get_embedding(text, model="text-embedding-3-small"):
   text = text.replace("\n", " ")
   return client.embeddings.create(input = [text], model=model).data[0].embedding

df['ada_embedding'] = df.combined.apply(lambda x: get_embedding(x, model='text-embedding-3-small'))
```

```python
Embeddings = df[["TA ID", "ada_embedding"]]
```

```python
response = client.embeddings.create(
    input="Machine Learning",
    model="text-embedding-3-small"
)

print(response.data[0].embedding)
Query = response.data[0].embedding
```

```python
from sklearn.metrics.pairwise import cosine_similarity
Embeddings["Similarity_Score"] = Embeddings["ada_embedding"].apply(lambda x: cosine_similarity([x], [Query])[0][0])
```

```python
Result = Embeddings.sort_values(by='Similarity_Score', ascending=False)["TA ID"].tolist()
```
