```python
import pandas as pd
import time
from sentence_transformers import SentenceTransformer, util
from sklearn.metrics.pairwise import cosine_similarity
```

```python
df = pd.read_csv('/content/Book1.csv')
df["combined"] = df.apply(lambda row: ' '.join([str(row[col]) for col in df.columns if col not in ['TA_ID','Area_of_Specialisation','Primary_Skills','Patents','Publications']]), axis=1)
print(df)
```

```python
def model():
  df = pd.read_csv('/content/Book1.csv')
  df["combined"] = df.apply(lambda row: ' '.join([str(row[col]) for col in df.columns if col not in ['TA_ID','Area_of_Specialisation','Primary_Skills','Patents','Publications']]), axis=1)
  def get_embedding(text):
    model = SentenceTransformer("all-MiniLM-L6-v2")
    print(model.max_seq_length)
    text = str(text).replace("\n", " ")
    print(text)
    Embed = model.encode(text, normalize_embeddings=False)
    return Embed
  df['primary_embedding'] = df.Primary_Skills.apply(lambda x: get_embedding(x))
  df['ada_embedding'] = df.combined.apply(lambda x: get_embedding(x))
  df['specialisation_embedding'] = df.Area_of_Specialisation.apply(lambda x: get_embedding(x))
  df['Experience_Industry'] = df.Experience_Industry.apply(lambda x: get_embedding(x))
  Embeddings = df[["TA_ID", "ada_embedding","specialisation_embedding", "primary_embedding","Experience_Industry"]]
  return Embeddings
```

```python
Embeddings = model()
```

```python
Embeddings
```

```python
def predict(text,Embeddings):
  model = SentenceTransformer("all-MiniLM-L6-v2")
  print(model.max_seq_length)
  text = str(text).replace("\n", " ")
  Query = model.encode(text, normalize_embeddings=False)
  Embeddings["Similarity_Score"] = (Embeddings["ada_embedding"].apply(lambda x: cosine_similarity([x], [Query])[0][0])*3)+((Embeddings["specialisation_embedding"].apply(lambda x: cosine_similarity([x], [Query])[0][0]))*4)+((Embeddings["primary_embedding"].apply(lambda x: cosine_similarity([x], [Query])[0][0]))*3)+((Embeddings["Experience_Industry"].apply(lambda x: cosine_similarity([x], [Query])[0][0]))*2)
  Result = Embeddings.sort_values(by='Similarity_Score', ascending=False)["TA_ID"].tolist()
  print(Result)
```

```python
predict("Machine Learning",Embeddings)
```
